```python
from DynamicalSystems import *
```

This package revolves around the long term stability of iterated maps. An iterated map is a sequence

\begin{equation}
x_{n+1} = f(r, x_n)
\end{equation}

for some parameter $r$. There are 3 important plots that tell us about the stability of an initial point $x_0$  for some function and parameter. The iteration plot shows how $x_n$ settles down (or doesn't). 

The cobweb plot shows the exact track of iteration over the function. From $x_0$ to $f(r, x_0)$ and so on

Lastly the orbit plot shows the stable value (if it exists) against the parameter that generated it. Orbit plots are great for seeing how and when chaotic behaviour becomes apperant in the system.



The prototypical example is the logistic function (for more info look at https://www.youtube.com/watch?v=ovJcsL7vyrk or https://en.wikipedia.org/wiki/Bifurcation_diagram )


\begin{equation}
 f(r, x) = r x (1-x)
\end{equation}

Which exhibits quick stability and nice cobweb plots for low values of $r$ and strange chaotic behaviour for high values of $r$. Lets start by defining the function.

```python
def f(r,x):
    return r*x*(1-x)
```

```python
for r in [1.5, 2.5, 3, 4]:
    iterate(f, 0.5, r, 1e-5, 40) 
```

We can see that as $r$ increases, the system stops stabilizing to a single point but instead oscilates between two points. In the last example there seems to be no stability at all. We can use the cobweb plots to understand whats happening here

```python
for r in [1.5, 2.5, 3, 4]:
    def g(x): return f(r, x)
    cweb(g)
```

Now in the first two examples we can see that the zigzagging line eventually selles on a point. Whereas in the third plot it bounces around between two points in a square. In the last example there is no stability whatsoever. The zigzag never settles down. So for what values of $r$ is the stability (even cyclical stability?) The orbit diagram will tell us:

```python
orbit(f, 0.5, 45, 100)
```

So at $r = 3$ we move from one stable point to two as we saw in our diagrams. And at $r = 3.5$ there is total chaos. Now the benefit of this program is that there is no limit on what functions we can place in here. Here are some examples:

\begin{equation}
x e^{r(1-x)} \\
 e^{-rx} \\
 r \cos x \\
 r \tan x
\end{equation}

```python
def f(r,x):
    return x*np.exp(r*(1-x))

orbit(f, 0.5, 100, 5, r_min = -1, r_max = 6, x_min = 0, x_max = 3)
```

```python
def f(r,x):
    return np.exp(-r*x)

orbit(f, 0.5, 100, 5, r_min = -1, r_max = 6, x_min = 0, x_max = 3)
```

```python
def f(r,x):
    return r*np.cos(x)

orbit(f, 0.5, 100, 5, r_min = -4, r_max = 4, x_min = -10, x_max = 10)
```

```python
def f(r,x):
    return r*np.tan(x)

orbit(f, 1, 100, 5, r_min = -2, r_max = 2, x_min = -10, x_max = 10)
```

Additionally this gives us the ability to see how stable a point is. Consider $g(x) = 3x - x^3$

```python
def g(x):
    return 3*x - x**3

cweb(g, xmin = -2.15, xmax = 2.15, x0 = 1.9, iterations = 10)
cweb(g, xmin = -2.15, xmax = 2.15, x0 = 2, iterations = 10)
cweb(g, xmin = -2.15, xmax = 2.15, x0 = 2.1, iterations = 2)
```

Note that even though the point $x_0$ has settled down into a cycle but the points to its immediate left and right are unstable. So $x_0$ has a weak kind of stability.
